## Malado Sissoko – Data Modeling

I worked on the data modeling section of the project. The purpose of my section was to test whether transaction features could help predict if a transaction would succeed or fail, and to identify patterns that could be useful in a real-world fraud or risk detection setting.

Before modeling, I cleaned feature values such as city names and card types so spelling differences would not be treated as separate categories. I also converted the transaction time into an hour variable so the models could measure time-based behavior.

The main features used were:

- Transaction amount  
- Hour of transaction  
- City  
- Card type  

I then trained and tested two machine learning models:

- Decision Tree  
- Random Forest  

Both models produced accuracy close to 50%. This means the selected variables alone were not strong enough for highly accurate prediction, but they were still useful for exploring patterns in transaction behavior.

In addition to the models, I tested relationships between variables and transaction outcomes. A p-value helps measure whether an observed relationship is likely meaningful or could have happened by random chance. Transaction amount showed the strongest measurable relationship with success or failure, while hour, city, and card type showed weaker relationships.

One important takeaway is that weak prediction accuracy does not mean the analysis had no value. In real-world data science, results can still help show which variables matter, which variables need improvement, and what additional data may be needed in future models.
