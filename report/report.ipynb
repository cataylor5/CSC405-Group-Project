### What are we trying to accomplish?
Something that is very inconsistent among the data given is the drastic spread of currency involved with each transaction. ### Why we should consider transactions of $0?
When looking at the data set there are some transactions which have no currency exchange amount. Naive intuition might suggest removing this transaction but when we are also considering recurring transactions and trying to trace frequency along with currency exchange amount the attempt of a transaction is still relavent in our data set.

### What data should be cleaned?
Some of the data that should be cleaned, removed, and modified are values which are missing entire cells of information such as a missing time or missing transaction type. Some of the other things that we want to look out for in our data are mispelled or "inconsistent" naming convention like the use if 'fail' and 'failed' which could cause problems when processing. Spelling errors are present in more than just the output something like 'visa' and 'vsa' are also present in our data set. These will come up as seperate data points in our output since the label Visa does not match vsa. Something to notice is the use of spaces and other characters like 'Master Card' compared to 'MasterCard' and 'VISA' compared to 'Visa'.

### Outliers in the data set
Outliers in this data set are commonly transaction amounts that are either very large or very small (even negative). Throughout our data cleaning and processing it is important to run our tests with a truncated more concentrated set of data as well as a fuzzier more general set of data. We cannot know what data we find important until we can know what question we are trying to answer. Because of this we believe that operating on a flexible basis is very important. Being able to easily adjust what your data set range is will aid in quickly checking multiple different groups of data.
