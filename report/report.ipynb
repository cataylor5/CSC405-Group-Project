# Group thesis
Transaction amount differs depending on transaction location and time.

## Statistical Framing of the Thesis

To support this thesis, the project does not rely only on visual patterns or assumptions. We also test the relationship statistically. In this report, transaction amount is treated as the response variable, while transaction location and transaction time are treated as explanatory variables. This allows us to examine whether the differences in transaction amount across cities and hours are large enough to be considered statistically meaningful.

The statistical portion of the project uses the following hypotheses:

**Null Hypothesis (H₀):** Transaction amount does not differ depending on transaction location or time.

**Alternative Hypothesis (H₁):** Transaction amount differs depending on transaction location and time.

These hypotheses guide the ANOVA tests used later in the report. Since transaction amount is numerical and city/hour create multiple comparison groups, ANOVA gives us a formal way to test whether the observed differences across location and time are statistically significant.

## Statistical Analysis & Hypothesis Testing - Carlos Taylor

For my part of the project, I focused on testing one part of the group’s overall research question from a statistical standpoint. The main question I wanted to answer was whether transaction amount actually differs depending on transaction location and transaction time. This seemed like a reasonable relationship to test because location and time both sound like factors that could affect transaction behavior, but I did not want to rely only on assumptions or visuals. I wanted to use a formal hypothesis test to see whether the data supported that idea.

### Research Question

Does transaction amount differ depending on transaction location and time?

### Hypotheses

**Null Hypothesis (H₀):** Transaction amount does not differ depending on transaction location or time.

**Alternative Hypothesis (H₁):** Transaction amount differs depending on transaction location and time.

In this part of the analysis, the response variable was `amount`, because that is the value being measured and compared. The explanatory variables were `city` and `hour`, because those represent the location and time factors being tested.

For both tests, I used a significance level of 0.05, meaning that p-values below 0.05 would be treated as statistically significant.

### Data Preparation for the Statistical Test

Before running the statistical tests, I prepared a separate statistical analysis dataframe so that I would not accidentally overwrite the original cleaned dataset used by the rest of the group. I cleaned the city values by making them lowercase, removing extra spaces, and correcting obvious spelling variations such as `tehr@n`, `thr`, and `thran` into `tehran`. I also converted the `time` column into a datetime format and extracted the hour of each transaction.

Rows were removed only if they were missing one of the specific variables needed for this test: transaction amount, city, or hour. This left **9,884 rows** available for the statistical analysis. I did not want to remove unrelated data unnecessarily, so the cleaning for this section was limited to what was needed for this specific hypothesis test.

### Test Selection

I used **ANOVA** because I was comparing a numerical variable across multiple groups. Transaction amount is numerical, while city creates multiple location groups and hour creates multiple time groups. ANOVA was appropriate because it allowed me to test whether the average transaction amount differed across those groups.

I ran two ANOVA tests:

1. One ANOVA test comparing transaction amount across cities.
2. One ANOVA test comparing transaction amount across hours of the day.

### City ANOVA Results

The city ANOVA test compared transaction amount across the different city groups.

**City ANOVA Result:**

- F-statistic: **0.583**
- p-value: **0.771**

Because the p-value was much greater than 0.05, we failed to reject the null hypothesis for location. This means there was not enough statistical evidence to conclude that transaction amount significantly differed by city.

This does not prove that location has no effect at all. It only means that, based on this dataset and this test, the city groups did not show a statistically significant difference in transaction amount.

### Time ANOVA Results

The time ANOVA test compared transaction amount across the different hours of the day.

**Hour ANOVA Result:**

- F-statistic: **1.522**
- p-value: **0.052**

The hour p-value was much closer to 0.05 than the city result, but it was still slightly above the standard significance level. Because of that, we still failed to reject the null hypothesis for time at the 0.05 level.

This suggests that hour may be more worth exploring than city, but there was still not enough statistical evidence to say that transaction amount significantly differed by hour of day.

### Supporting Correlation Check

I also looked at correlation between transaction hour and transaction amount as a supporting check. The Pearson correlation was approximately **0.007**, with a p-value of approximately **0.488**. The Spearman correlation was approximately **0.019**, with a p-value of approximately **0.065**.

Both results were very weak, which supports the ANOVA result. In other words, transaction hour did not show a strong relationship with transaction amount in this dataset.

This was important because it showed why the graph alone was not enough and why the statistical test was needed.

### Statistical Analysis Conclusion

Overall, the statistical tests did not provide enough evidence to support the alternative hypothesis. The city ANOVA had a p-value of **0.771**, and the hour ANOVA had a p-value of **0.052**. Since both were greater than 0.05, we failed to reject the null hypothesis.

The main takeaway from this section is that location and time alone do not strongly explain transaction amount in this dataset. They may still be useful as supporting features in a larger model, but based on this analysis, they do not appear strong enough by themselves to explain major differences in transaction amount.

This is still a useful result because it shows that the project should not rely only on location and time to understand transaction amount. Other transaction features, such as status, card type, or other cleaned variables, may be more helpful for explaining patterns in the dataset.
